```python
sql_query = """
SELECT COUNT(id) as numListings, origin, product
FROM listings
WHERE (product LIKE '%fentanyl%'
    AND product LIKE '%X 10%')
GROUP BY product
ORDER BY 1 DESC
LIMIT 100 OFFSET 0;
"""

pd.read_sql_query(sql_query,con)
```
